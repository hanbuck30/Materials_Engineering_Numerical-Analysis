```python
import numpy as np
import matplotlib.pyplot as plt

#   
D = 1.0
N = 100

L = 7
dx = L / N
x = np.linspace(-3, 3, N)
C = np.zeros(N)  #    (0 )
C[(x >= -1) & (x <= 1)] = 1  #    (  1 )

#  
t = 0
dt = 0.0001
nsteps = 5000
for i in range(nsteps):
    #    
    dCdt = D * np.diff(C, 2) / dx ** 2
    #    (Dirichlet  )
    dCdt[0] = 0
    dCdt[-1] = 0
    # Euler    
    C[1:-1] += dt * dCdt
    #  
    t += dt

#  
plt.plot(x, C)
plt.xlabel('x')
plt.ylabel('Concentration')
plt.title('Anomalous diffusion with D = {} cm^2/s'.format(D))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#  
Vf = 0.5 #  
sigma_f = 400 #   (Fiber tensile strength)
sigma_m = 50 #   (Matrix tensile strength)
eta_0_list = [1.0, 0, 0.5, 3/8, 0.2] # ( : 1.0,  : 0,  : 0.5, 2 random  : 3/8, 3 random  : 0.2)
Lc = 10 #    (Fiber critical length)

#    
L_range = np.linspace(1, 100, 99) # 1 ~ 100   99  

#  


plt.figure(figsize=(7,7)) 
for i, eta_0 in enumerate(eta_0_list):
    #  
    eta_L = np.zeros(len(L_range))
    for j in range(len(L_range)):
        if L_range[j] >= Lc :
            eta_L[j] = 1 - Lc / (2 * L_range[j])
        else :
            eta_L[j] = L_range[j] / (2 * Lc)
    sigma_comp = eta_0 * eta_L * sigma_f * Vf + sigma_m * (1 - Vf)
    if i ==0:
        plt.plot(L_range,sigma_comp, color='limegreen' ,label =" : 1.0") 
    elif i ==1:
        plt.plot(L_range,sigma_comp, color='violet',label =" : 0") 
    elif i ==2:
        plt.plot(L_range,sigma_comp, color='dodgerblue',label =" : 0.5") 
    elif i ==3:
        plt.plot(L_range,sigma_comp,'r',label ="2 random  : 3/8") 
    elif i ==4:
        plt.plot(L_range,sigma_comp,'b',label ="3 random  : 0.2")   
plt.legend(loc=(0.85,0.85))
plt.xlabel('Fiber length')
plt.ylabel('Tensile strength (MPa)')       
plt.title('')
plt.ylim(0,250)

plt.show()
```
